Metodo shape[] : Devuelve el dato relacionado a la cantidad de filas y columnas de nuestro dataframe (confirma las dimensiones de nuestro dataframe)

```python
tabla.shape    #devuelve 460,7
tabla.shape[0] #devuelve 460
tabla.shape[1] #devuelve 7
```

Si queremos visualizar una cantidad especifica de filas usamos .head( indicando el valor a mostrar)

```python
tabla.head() #por defecto es 5
tabla.head(15) #Si queremos visualizar una cantidad de filas determinadas
```

```python
tabla.tail() #por defecto los ultimos 5 
tabla.tail(10)
```
