```python
import pandas as pd  # Importa pandas con alias 'pd'
tabla = pd.read_csv('/Users/miguelpantigoso/Downloads/ventasTotales.csv', sep=';', skiprows=5)
# Lee el archivo CSV, usando ";" como separador y saltando las primeras 5 filas

tabla.head() 
```

```python
tabla.groupby('Región')
```

```python
type(tabla.groupby('Región'))
```

```python
list(tabla.groupby('Región'))
```

```python
#iterar entre grupos
for group_key, group_value in tabla.groupby('Región'):
    print(group_key)
    print(group_value)
```

```python
tabla.groupby('Región').size()
```

```python
tabla.groupby(['Región','Vendedor','Ganancia']).agg(['min','max','count'])
```
