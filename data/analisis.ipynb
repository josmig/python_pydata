Value_counts o contar valores es un metodo que hace devolver un objeto de tipo serie que cuenta todos los valores unicos de nuestro dataframe

```python
import pandas as pd
tabla = pd.read_csv('/Users/miguelpantigoso/Downloads/ventasTotales.csv', sep=';', skiprows=5)
tabla.head()
```

```python
tabla.Vendedor.value_counts() #Con esto vemos cuantas veces en nuestro registro se repiten tal vendedor en este ejemplo quiere decir que 
#tal vendedor tuvo tantas cantidades de venta
tabla.Vendedor.value_counts(ascending=True) #asciende a al que mas tiene por defecto es false
tabla.Región.value_counts(ascending=True, dropna=False) #dropna es para eliminar los registros nulos por defecto es true
```

sort_values(ordenar los valores) ordena los valores de una serie de datos o dataframes, tiene parametros axis por defecto 0 y forma ascendente de manera predeterminada. Cuando se utiliza en un dataframe podremos ordenarlos ya sea de forma ascendente como descendente.

```python
tabla.Vendedor.sort_values() #en este caso aplicamos a una serie
tabla.head(30).sort_values(by=['Vendedor','Ganancia'])#aplicado a una dataframe y eligiendo a que columnas
```

Condiciones Booleanas :
Se utilizan vectores booleanos para filtrar información

```python
tabla.Artículo == 'Gorras' #muestra una serie de todos los registro de nuestra tabla donde se vendio tal producto
tabla[tabla.Artículo == 'Gorras'] #mas ordenado
#Si queremos visualizar archivos que cumplen dos o mas condiciones
tabla[(tabla.Artículo=='Gorras') & (tabla.Región =='Norte')]
```

Manejo de Strings : para poder manipular en la busqueda manipulando si contiene tal caracter letra o numero etc

```python
tabla[tabla.Vendedor.str.contains('Trevor')] #con esto buscamos un nombre en especial todos lo que incluya
tabla[tabla.Vendedor.str.startswith('D')] #con esto buscamos lo que empieza con T
```
