# Graficos basicos

```python
import pandas as pd                # Importa pandas con alias 'pd'
import matplotlib.pyplot as plt    # Importa Matplotlib para gráficos
%matplotlib inline                 
# (Solo en Jupyter): Muestra los gráficos en el mismo notebook

tabla = pd.read_csv('/Users/miguelpantigoso/Downloads/ventasTotales.csv', sep=';', skiprows=5)
# Lee el archivo CSV, usando ";" como separador y saltando las primeras 5 filas

tabla.head()                       # Muestra las primeras 5 filas del DataFrame
```

```python
ali = tabla[tabla.Categoría == 'Comestibles']
ali.head()
```

```python
ali.Artículo.value_counts().plot(kind='line') #podemos poenr un punto y como al final de este codigo apra que salga eso de axes etc
ali.Artículo.value_counts().plot(kind='line',figsize=(10,5))
```

## Graficos de Barras

```python
ali.Artículo.value_counts().plot(kind='bar'); #podemos poenr un punto y como al final de este codigo apra que salga eso de axes etc
```

## Grafico Barra horizontal

```python
ali.Artículo.value_counts().plot(kind='barh', color='red') #podemos poenr un punto y como al final de este codigo apra que salga eso de axes etc
```

## Grafico circular o de torta

```python
ali.Artículo.value_counts().plot(kind='pie') #podemos poenr un punto y como al final de este codigo apra que salga eso de axes etc
ali.Artículo.value_counts().plot(kind='pie', colormap='ocean')
```

## Grafico de dispersión

```python
tabla.plot(kind='scatter', x='Unidades', y='Ganancia'); #podemos poenr un punto y como al final de este codigo apra que salga eso de axes etc
```
